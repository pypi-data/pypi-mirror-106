### Chargement des données pour tester

```python
# Chargement de données d'une enquête (Image de la Science, 2011)
import pyreadstat
data, meta = pyreadstat.read_sav('bdd2011.sav')

# Fonctions spécifiques de manipulation des métadonnées
def question(v,meta):
    return meta2011.column_names_to_labels[v]
def modalites(v,meta):
    return meta.value_labels[meta.variable_to_label[v]]

# Recodage de quelques variables pour le test
data["interetscience"] = data['q2'].replace(modalites('q2',meta))
data["apportscience"] = data['q15'].replace(modalites('q15',meta))
data["sexe"] = data["rs1"].replace(modalites("rs1",meta))
data["age"] = data['rs2']
data["apportsciencepositif"] = data["apportscience"].apply(lambda x : 1 if x == "Plus de bien que de mal" else 0)
```

### Tri à plat pondéré

```python
pyshsdev.tri_a_plat(data,"interetscience","poids1")
```

### Tableau croisé pondéré

```python
pyshsdev.tableau_croise(data,"interetscience","sexe","poids1")
```

### Tableau croisé multiple pondéré

```python
pyshsdev.tableau_croise_multiple(data,"apportscience",
            {"sexe":"Sexe","interetscience":"Intérêt Science"},"poids1")
```

### Régression logistique rapide

```python
pyshsdev.regression_logistique(data,"apportsciencepositif",
                    {"sexe":"Sexe","age":"Age","interetscience":"Intérêt pour la science"},
                    table_only=True)
```
